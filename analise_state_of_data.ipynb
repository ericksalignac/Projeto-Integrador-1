```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Carregamento dos dados
url = 'State_of_data_BR_2023_Kaggle - df_survey_2023.csv'
df = pd.read_csv(url)
```

```python
# Limpar os nomes das colunas
df.columns = df.columns.str.strip()  # Remove espaços em branco
df.columns = df.columns.str.lower()  # Converte para minúsculas
df.columns = df.columns.str.replace(' ', '_')  # Substitui espaços por underscores
df.columns = df.columns.str.replace(r'[^a-zA-Z0-9_]', '', regex=True)  # Remove caracteres especiais

# Formatando os nomes das colunas como solicitado
formatted_columns = [f'"{col}"' for col in df.columns]
formatted_columns = "\n".join([f"[{i+1}] {col}" for i, col in enumerate(formatted_columns)])

print(formatted_columns)
```

```python
# Selecionando e renomeando colunas
dados_selecionados = df.rename(columns={
    'p0_id': 'id',
    'p1_a__idade': 'idade',
    'p1_a_1__faixa_idade': 'faixa_idade',
    'p1_b__genero': 'genero',
    'p1_c__corracaetnia': 'cor_raca_etnia',
    'p1_d__pcd': 'pcd',
    'p1_g__vive_no_brasil': 'vive_no_brasil',
    'p1_i__estado_onde_mora': 'estado_onde_mora',
    'p2_i__quanto_tempo_de_experincia_na_rea_de_dados_voc_tem': 'experiencia_dados',
    'p2_h__faixa_salarial': 'faixa_salarial',
    'p2_o_1__remuneraosalrio': 'remuneracao',
    'p4_a_1__atuacao': 'atuacao'
})[[
    'id',
    'idade',
    'faixa_idade',
    'genero',
    'cor_raca_etnia',
    'pcd',
    'vive_no_brasil',
    'estado_onde_mora',
    'experiencia_dados',
    'faixa_salarial',
    'remuneracao',
    'atuacao'
]]

# Tratamento de dados ausentes ou formatos específicos
dados_selecionados['vive_no_brasil'] = dados_selecionados['vive_no_brasil'].astype(bool)
dados_selecionados['faixa_idade'] = dados_selecionados['faixa_idade'].astype('category')
dados_selecionados['genero'] = dados_selecionados['genero'].astype('category')
dados_selecionados['cor_raca_etnia'] = dados_selecionados['cor_raca_etnia'].astype('category')
dados_selecionados['pcd'] = dados_selecionados['pcd'].astype('category')
dados_selecionados['atuacao'] = dados_selecionados['atuacao'].astype('category')

# Exibindo as primeiras linhas do DataFrame
dados_selecionados.head()
```

```python
import re
import numpy as np
# Função para extrair faixa salarial
def extract_faixa_salarial(salario_str):
    if pd.isna(salario_str):
        return np.nan
    # Extrair os valores numéricos da faixa salarial
    numeros = re.findall(r'\d+', salario_str.replace('.', ''))
    if len(numeros) == 2:
        return (int(numeros[0]) + int(numeros[1])) / 2
    elif len(numeros) == 1:
        return int(numeros[0])
    else:
        return np.nan

# Transformações e conversões
dados_tratados = dados_selecionados.copy()

# Adicionar coluna experiencia_dados_num com base nas condições especificadas
dados_tratados['experiencia_dados_num'] = dados_tratados['experiencia_dados'].map({
    "Menos de 1 ano": 0.5,
    "de 1 a 2 anos": 1.5,
    "de 3 a 4 anos": 3.5,
    "de 4 a 6 anos": 5,
    "de 6 a 8 anos": 7,
    "Mais de 10 anos": 10
})

# Adicionar coluna faixa_salarial_num usando a função extract_faixa_salarial
dados_tratados['faixa_salarial_num'] = dados_tratados['faixa_salarial'].apply(extract_faixa_salarial)

# Remover valores nulos nas colunas experiencia_dados_num e faixa_salarial_num
dados_tratados = dados_tratados.dropna(subset=['experiencia_dados_num', 'faixa_salarial_num'])

# Exibindo as primeiras linhas do DataFrame tratado
dados_tratados.head()
```
